```python
def enumerate_alt(xs, start=0, step=1):
    for x in xs:
        yield (start, x)
        start += step


imgs_dir = 'CholecT50/videos'

with open("data_cholecT50.json", "w") as file:
    img_1_paths_train = []
    img_2_paths_train = []
    labels_train = []
    img_1_paths_test = []
    img_2_paths_test = []
    labels_test = []
    for num, patient in enumerate(os.listdir(imgs_dir)):
        patient_directory = os.path.join(imgs_dir, patient)
        directory = os.listdir(patient_directory)
        filtered_dir = directory[::50]
        print(len(filtered_dir), len(directory))
        for i, picture_1_path in enumerate(filtered_dir):
            #picture_1 = Image.open(picture_1_path)
            for j, picture_2_path in enumerate_alt(filtered_dir):
                if i == j:
                    pass
                elif (".jpg" not in picture_1_path and '.png' not in picture_1_path) or (".jpg" not in picture_2_path and '.png' not in picture_2_path):
                    pass
                else:
                    #picture_2 = Image.open(picture_2_path)
                    label = 1 if j > i else 0
                    if num > 4:
                        img_1_paths_train.append(os.path.join(patient_directory, picture_1_path))
                        img_2_paths_train.append(os.path.join(patient_directory, picture_2_path))
                        labels_train.append(label)
                    else:
                        img_1_paths_test.append(os.path.join(patient_directory, picture_1_path))
                        img_2_paths_test.append(os.path.join(patient_directory, picture_2_path))
                        labels_test.append(label)
                    #dictionaries.append(dictionary)
        print('finished video')
        print(len(img_1_paths_train))
    zipped_train = list(zip(img_1_paths_train, img_2_paths_train, labels_train))
    zipped_test = list(zip(img_1_paths_test, img_2_paths_test, labels_test))
    random.shuffle(zipped_train)
    random.shuffle(zipped_test)
    img_1_paths_train, img_2_paths_train, labels_train = zip(*zipped_train)
    img_1_paths_test, img_2_paths_test, labels_test = zip(*zipped_test)
    dictionary_train = {"img_path_1": img_1_paths_train, "img_path_2": img_2_paths_train, "labels": labels_train}
    dictionary_test = {"img_path_1": img_1_paths_test, "img_path_2": img_2_paths_test, "labels": labels_test}
    combined_dicitonary = {"train": dictionary_train, "test": dictionary_test}
    json.dump(combined_dicitonary, file)
                    #file.write(f"{os.path.join(patient_directory, picture_1_path)}, {os.path.join(patient_directory, picture_2_path)}, {label}\n")
                    #input_images = (picture_1, picture_2)
```
